```python
#IMPORT THE LIBRARIES
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# COVID 19 ANALYSIS
#QUE: IS THERE ANY RELATIONSHIP BETWEEN SPREAD OF CORONA VIRUS IN OUR COUNTRY AND HOW HAPPY THE PEOPLE ARE LIVING IN THAT PARTICULAR COUNTR
```

```python
#import the dataset
dataset = pd.read_csv("C:\\Users\\USER\\Documents\\data science\\COVID 19 ANALYSIS\\covid19_confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
```

```python
#delete useless columns
df = dataset.drop(["Lat","Long"], axis = 1 , inplace = True)
```

```python
dataset.head()
```

```python
#aggregate the rows by the country
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
#visualize the data related to a country
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
#range of covid cases of china for the first 3 days
corona_dataset_aggregated.loc["China"][:3].plot()
```

```python
#Calculate the first derivative of the curve
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
#maximum infection rate
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregated
```

```python
#create new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
```

```python
corona_data
```

```python
#importing the happiness report
happiness_report = pd.read_csv("C:\\Users\\USER\\Documents\\data science\\COVID 19 ANALYSIS\\worldwide_happiness_report.csv")
```

```python
happiness_report
```

```python
#delete the useless columns
useless_cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness_report.drop(useless_cols, axis=1, inplace = True)
```

```python
happiness_report.head()
```

```python
happiness_report.set_index("Country or region", inplace = True)
happiness_report.head()
```

```python
corona_data.shape
```

```python
#joining two datasets
data = corona_data.join(happiness_report, how = "inner")
```

```python
#visualization
x = data["GDP per capita"]
y = data["Max_infection_rates"]
sns.scatterplot(x=x,y=np.log(y))
sns.regplot(x=x,y=np.log(y))
plt.xlabel('GDP per capita')
plt.ylabel('Log Max Infection Rates')
plt.title('Scatter plot with Log Max Infection Rates')
plt.show()
```

```python
x = data["Social support"]
y = data["Max_infection_rates"]
sns.scatterplot(x=x,y=np.log(y))
sns.regplot(x=x,y=np.log(y))
plt.xlabel('Social support')
plt.ylabel('Log Max Infection Rates')
plt.title('Scatter plot with Log Max Infection Rates')
plt.show()
```
